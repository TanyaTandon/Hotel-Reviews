```python
from itertools import chain
import numpy as np
import pandas as pd
```

```python
hotel = pd.read_csv('../00Data/Hotel_Reviews_v2.csv')
```

```python
hotel.head(10)
```

```python
hotel['Tags']=hotel["Tags"].str.strip('[]')
hotel['Tags'] = hotel['Tags'].str.replace('\' ','')
hotel['Tags'] = hotel['Tags'].str.replace(' \'','')
hotel['Tags'] = hotel['Tags'].str.replace(', ',',')
hotel['Tags'] = hotel['Tags'].str.replace(' ','_')
hotel['Tags'] = hotel['Tags'].str.lower()
```

```python
hotel['Tags_U'] = hotel['Tags'].str.split(',')
```

```python
hotel.head(10)
```

```python
id_tag = pd.DataFrame({ "ID": np.repeat(hotel.ID.values, [len(x) for x in hotel.Tags_U]),"Tag": list(chain.from_iterable(hotel.Tags_U))})

```

```python
keys = pd.read_csv('key.csv')
```

```python
id_tag_2 = id_tag.merge(keys, left_on = 'Tag', right_on = 'key')
```

```python
id_tag_2
```

```python
id_tag_2['key_type'] = id_tag_2['key_type'].astype(str) + "_cust"
```

```python
id_level = id_tag_2.groupby(['ID','key_type']).agg({'key_update':'first'}).unstack('key_type')
```

```python
id_level.head(10)
```

```python
id_level.fillna("none").to_csv("id_tags.csv")
```
