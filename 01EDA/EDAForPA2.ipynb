```python
import math
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
def get_doc(df):
    
    doc_info = []
    i = 0
    for i in range(len(df)):
        count = df['Tags'][i].count(' ') + 1
        temp = {'hotel_id': df['Hotel_Address'][i], 'doc_length': count}
        doc_info.append(temp)
    return doc_info
```

```python
def create_freq_dict(df):
    
    i=0
    freqDict_list = []
    for i in range(len(df)):
        freq_dict = {}
        words = word_tokenize(df['Tags'][i])
        for word in words:
            word = word.lower()
            if word in freq_dict:
                freq_dict[word] += 1
            else:
                freq_dict[word] = 1
            temp = {'hotel_id': df['Hotel_Address'][i], 'freq_dict': freq_dict}
        freqDict_list.append(temp)
    return freqDict_list
```

```python
def computeTF(doc_info, freqDict_list):
    
    TF_scores = []
    for temp_dict in freqDict_list:
        id = temp_dict['hotel_id']
        den = [element for element in doc_info if element['hotel_id'] == id][0].get('doc_length')
        for k in temp_dict['freq_dict']:
            temp = {'hotel_id':id,
                   'TF_score': temp_dict['freq_dict'][k]/den,
                    'key': k}
            TF_scores.append(temp)
    return TF_scores
```

```python
def computeIDF(doc_info, freqDict_list):
    
    IDF_scores = []
    for temp_dict in freqDict_list:
        id = temp_dict['hotel_id']
        for k in temp_dict['freq_dict'].keys():
            count = sum([k in tempDict['freq_dict'] for tempDict in freqDict_list])
            temp = {'hotel_id':id,
                   'IDF_score': math.log(len(doc_info)/count),
                    'key': k}
            IDF_scores.append(temp)
    return IDF_scores
```

```python
def computeTFIDF(TF_scores, IDF_scores):
    
    TFIDF_scores = []
    for i in range(len(IDF_scores)):
        id = IDF_scores[i]['hotel_id']
        k = IDF_scores[i]['key']
        tf_idf = IDF_scores[i]['IDF_score'] * TF_scores[i]['TF_score']
        temp = {'hotel_id':id,
                'TFIDF_score': tf_idf,
                'key': k}
        TFIDF_scores.append(temp)
    return TFIDF_scores
```

```python
import time
t0 = time.time()
doc = get_doc(hotel_tags)
freqDict = create_freq_dict(hotel_tags)
t1 = time.time()

t1-t0
```

```python
t0 = time.time()
tf_scores = computeTF(doc, freqDict)
idf_scores = computeIDF(doc, freqDict)
t1 = time.time()

t1-t0
```

```python
t0 = time.time()
tfidf_scores = computeTFIDF(tf_scores, idf_scores)
t1 = time.time()

t1-t0
```
